```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
## Define functions for text cleaning and preprocessing
```

```python
def clean_text(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove special characters and digits
    text = re.sub('[^a-zA-Z]', ' ', text)
    
    # Remove extra whitespaces
    text = re.sub('\s+', ' ', text)
    
    return text
```

```python
text_columns = ['job_post', 'job_location', 'job_description']

for column in text_columns:
    # Clean text
    df[column] = df[column].apply(clean_text)
```

```python
df.head()
```

```python
def remove_stopwords(text):
    # Tokenize text
    tokens = text.split()
    
    # Remove stopwords
    stopwords_list = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stopwords_list]
    
    # Join filtered tokens back into text
    filtered_text = ' '.join(filtered_tokens)
    
    return filtered_text
```

```python
for column in text_columns:
    # Remove stopwords
    df[column] = df[column].apply(remove_stopwords)
```

```python
# Define the categorical columns
categorical_columns = ['exp_required', 'salary_offered']

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Print the encoded DataFrame
df_encoded.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Get the list of numerical column names
numerical_columns = df_encoded.select_dtypes(include='number').columns.tolist()

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Apply Min-Max Scaling to the numerical columns
df_encoded[numerical_columns] = scaler.fit_transform(df_encoded[numerical_columns])

# Display the normalized dataset
df_encoded.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Perform EDA on categorical variables
categorical_columns = ['exp_required', 'salary_offered']

# Plot the count of each category in categorical columns
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f"Count of {column}")
    plt.xticks(rotation=45)
    plt.show()

# Perform EDA on numerical variables
numerical_columns = ['company_rating']

# Plot the distribution of numerical columns
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df_encoded, x=column, kde=True)
    plt.title(f"Distribution of {column}")
    plt.show()

# Perform correlation analysis
plt.figure(figsize=(10, 8))
sns.heatmap(df_encoded.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
